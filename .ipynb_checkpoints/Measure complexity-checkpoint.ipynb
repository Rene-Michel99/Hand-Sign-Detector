```python
import random
import matplotlib.pyplot as plt
import time

def quest(n):
    vetor = []

    for i in range(n):
        vetor.append(random.randint(0,5*n))

    dic = {}
    for i in vetor:
        if i not in dic.keys():
            dic[i] = 1
        else:
            dic[i] += 1

def bubble_sort(n):
    for i in range(len(n)):
        for j in range(len(n)):
            if i!=j and n[i] > n[j]:
                aux = n[i]
                n[i] = n[j]
                n[j] = aux
                
def third(n):
    vetor = []

    for i in range(n):
        vetor.append(random.randint(1,3))

    vetb = [0 for i in range(5*n)]

    for i in vetor:
        vetb[i] += 1
    print(vetor)
    for i,qntd in enumerate(vetb):
        if qntd > 3:
            pass
            
def measure(func,args):
    start_time = time.time()
    func(args)
    end_time = time.time()
    return end_time - start_time
```

```python
args = [i*300 for i in range(5)]
times = []
for i in args:
    vec = [0 for i in range(i)]
    times.append(measure(bubble_sort,vec))
    
print(times)
```

```python
plt.plot(args, times, 'b')
plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Quadratic Complexity')
plt.show()
```

```python
ns = [50,100,1500,20000]
times = []
for i in ns:
    times.append(measure(third,i))
    
print(times)
```

```python
plt.plot(ns, times, 'b')
plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Linear Complexity')
plt.show()
```
